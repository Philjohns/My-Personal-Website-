```python
def distort_landmarks(landmarks, identity_idx, neutral_idx, intensity=5):
    """
    Distorts selected facial landmarks.

    Parameters:
        landmarks: list of (x, y) coordinates
        identity_idx: indices to distort more (identity features)
        neutral_idx: indices to distort slightly (optional)
        intensity: maximum pixel shift for distortion (default 5)

    Returns:
        new_landmarks: list of new (x, y) coordinates
    """
    new_landmarks = []

    for i, (x, y) in enumerate(landmarks):
        if i in identity_idx:
            dx = np.random.normal(0, intensity)
            dy = np.random.normal(0, intensity)
        elif i in neutral_idx:
            dx = np.random.normal(0, intensity * 0.5)
            dy = np.random.normal(0, intensity * 0.5)
        else:
            dx = dy = 0  # No change for expression landmarks
        new_landmarks.append((x + dx, y + dy))

    return new_landmarks
```

```python
import scipy.spatial

def get_triangles(landmarks, w, h):
    """
    Perform Delaunay triangulation and return triangle vertex indices.
    """
    # Convert landmarks to 2D array
    points = np.array(landmarks, dtype=np.float32)
    subdiv = scipy.spatial.Delaunay(points)
    return subdiv.simplices  # indices of triangle points
```

```python
def warp_triangle(img_src, img_dst, tri_src, tri_dst):
    """
    Warp one triangle from source image to destination image using affine transform.
    """
    # Bounding rectangles
    r1 = cv2.boundingRect(np.float32([tri_src]))
    r2 = cv2.boundingRect(np.float32([tri_dst]))

    # Offset points to local rectangle space
    tri_src_rect = []
    tri_dst_rect = []
    for i in range(3):
        tri_src_rect.append(((tri_src[i][0] - r1[0]), (tri_src[i][1] - r1[1])))
        tri_dst_rect.append(((tri_dst[i][0] - r2[0]), (tri_dst[i][1] - r2[1])))

    # Mask for triangle
    mask = np.zeros((r2[3], r2[2], 3), dtype=np.uint8)
    cv2.fillConvexPoly(mask, np.int32(tri_dst_rect), (1.0, 1.0, 1.0), 16, 0)

    # Extract triangle from source
    img1_rect = img_src[r1[1]:r1[1]+r1[3], r1[0]:r1[0]+r1[2]]
    size = (r2[2], r2[3])

    # Affine transform
    mat = cv2.getAffineTransform(np.float32(tri_src_rect), np.float32(tri_dst_rect))
    warped_triangle = cv2.warpAffine(img1_rect, mat, size, flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REFLECT_101)

    # Place the warped triangle into the destination image
    img_dst[r2[1]:r2[1]+r2[3], r2[0]:r2[0]+r2[2]] = \
        img_dst[r2[1]:r2[1]+r2[3], r2[0]:r2[0]+r2[2]] * (1 - mask) + warped_triangle * mask
```

```python
# Prepare output canvas
warped_img = np.zeros_like(image_rgb)

# Get triangles using original landmark positions
triangles = get_triangles(original_landmarks, w, h)

# Warp each triangle
for tri_indices in triangles:
    tri_src = [original_landmarks[i] for i in tri_indices]
    tri_dst = [distorted_landmarks[i] for i in tri_indices]
    warp_triangle(image_rgb, warped_img, tri_src, tri_dst)

# Show final output
plt.figure(figsize=(6, 6))
plt.title("Warped Face Image (Distorted Identity)")
plt.imshow(np.clip(warped_img, 0, 255).astype(np.uint8))
plt.axis('off')
plt.show()
```

```python

import gradio as gr
import cv2
import numpy as np
import mediapipe as mp
from PIL import Image
from deepface import DeepFace
from scipy.interpolate import Rbf

# Mediapipe initialization
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1)

identity_landmarks = list(range(0, 17)) + list(range(65, 106)) + list(range(295, 335))
expression_landmarks = list(range(61, 91)) + list(range(146, 178)) + list(range(181, 211)) + list(range(234, 263))
neutral_landmarks = [i for i in range(468) if i not in identity_landmarks and i not in expression_landmarks]

def distort_landmarks(landmarks, identity_idx, intensity=2):
    new_landmarks = []
    for i, (x, y) in enumerate(landmarks):
        if i in identity_idx:
            dx = np.random.normal(0, intensity)
            dy = np.random.normal(0, intensity)
        else:
            dx = dy = 0
        new_landmarks.append((x + dx, y + dy))
    return new_landmarks

def apply_tps_warp(image, src_pts, dst_pts):
    h, w = image.shape[:2]
    map_x = np.zeros((h, w), dtype=np.float32)
    map_y = np.zeros((h, w), dtype=np.float32)

    src_pts = np.array(src_pts)
    dst_pts = np.array(dst_pts)

    rbf_x = Rbf(dst_pts[:, 0], dst_pts[:, 1], src_pts[:, 0], function='thin_plate')
    rbf_y = Rbf(dst_pts[:, 0], dst_pts[:, 1], src_pts[:, 1], function='thin_plate')

    grid_x, grid_y = np.meshgrid(np.arange(w), np.arange(h))
    map_x = rbf_x(grid_x, grid_y).astype(np.float32)
    map_y = rbf_y(grid_x, grid_y).astype(np.float32)

    warped = cv2.remap(image, map_x, map_y, interpolation=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REFLECT_101)
    return warped

def process_image(input_img, distortion_level):
    image = np.array(input_img.convert("RGB"))
    h, w, _ = image.shape

    results = face_mesh.process(image)
    if not results.multi_face_landmarks:
        return image, image, "Face not detected", "Face not detected"

    original_landmarks = [(lm.x * w, lm.y * h) for lm in results.multi_face_landmarks[0].landmark]
    distorted_landmarks = distort_landmarks(original_landmarks, identity_landmarks, intensity=distortion_level)

    warped_img = apply_tps_warp(image, original_landmarks, distorted_landmarks)

    # Save temp images
    original_path = "original_tmp.jpg"
    warped_path = "warped_tmp.jpg"
    Image.fromarray(image).save(original_path)
    Image.fromarray(warped_img).save(warped_path)

    # Evaluate identity and emotion
    id_result = DeepFace.verify(original_path, warped_path, model_name='Facenet', enforce_detection=False)
    emotion_original = DeepFace.analyze(original_path, actions=["emotion"], enforce_detection=False)
    emotion_warped = DeepFace.analyze(warped_path, actions=["emotion"], enforce_detection=False)

    return image, warped_img, f"{id_result['distance']:.3f}", f"{emotion_original[0]['dominant_emotion']} ➜ {emotion_warped[0]['dominant_emotion']}"

interface = gr.Interface(
    fn=process_image,
    inputs=[
        gr.Image(type="pil", label="Upload Face Image"),
        gr.Slider(minimum=1, maximum=4, step=1, value=2, label="Distortion Level (Subtle)")
    ],
    outputs=[
        gr.Image(label="Original"),
        gr.Image(label="Warped (TPS Natural Look)"),
        gr.Text(label="Identity Distance (Cosine)"),
        gr.Text(label="Emotion Change")
    ],
    title="TPS-Based Identity Distortion",
    description="Thin Plate Spline warping for natural-looking identity alteration. Subtle changes preserve appearance while reducing face match accuracy."
)

interface.launch(share=True)

```
